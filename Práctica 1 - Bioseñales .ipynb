```python
#1 Creación de vectores
a = np.array([3.1,1, -0.5, -3.2, 6])
b= np.array([1, 3, 2.2, 5.1, 1])

#2 Producto escalar de los vectores
escalar= np.dot(a,b)
print(escalar)
```

```python
#4 Creación de una matriz definida
A=np.array([(2,-1,-3),(4,1.5,-2.5),(7.3, -0.9, 0.2)])
print(A)
```

```python
#5 Transpuesta de la matriz definida
transf=A.T
print(transf)
```

```python
# 10. Construir la función 𝑦[𝑛] = sin (𝜋 ∗ 0.12𝑛) en el intervalo 0 <= n <= 100

n = np.arange(0,101)

sin = np.sin(np.pi*0.12*n)

# 11. Construir otra senoidal 𝑦2[𝑛] = cos(2𝜋 ∗ 0.03𝑛)

cos = np.cos(2*np.pi*0.03*n)

# 12. Hacer la suma y el producto de las dos señales anteriores

suma = sin+cos
producto = sin*cos

# 13. Graficar las señales seno y coseno en la misma figura

plt.figure(figsize=(10,6))
plt.plot(n,sin,label='Señal seno')
plt.plot(n,cos,label='Señal coseno')
plt.title('Señales seno y coseno')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.legend()
```

```python
# 13. Graficar la suma y el producto de las señales seno y coseno en la misma figura

plt.figure(figsize=(10,6))
plt.plot(n,suma,label='Suma')
plt.plot(n,producto,label='Producto')
plt.title('Suma y producto')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.legend()
```

```python
df=pd.read_csv('pacientes_VitalDB (1).csv')
```

```python
print("Primeras filas: \n", df.head()) #Mostrar primeras filas del dataframe
print("\nÚltimas filas: \n",df.tail()) #Mostrar últimas filas del dataframe
```

```python
df["BMI"] = df["height"] / ((df["weight"] / 100) ** 2)
print("\nÚltimas filas: ")
print(df.tail())
```

```python
def clasificacion(bmi):
    if bmi < 18.5:
        return "Bajo peso"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

df["Categoria IMC"] = df["BMI"].apply(clasificacion)

print(df)
```
